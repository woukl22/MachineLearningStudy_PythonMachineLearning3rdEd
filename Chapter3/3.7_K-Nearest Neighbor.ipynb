# 3.7 k-최근접 이웃: 게으른 학습 알고리즘

#### **k-최근접 이웃(K-Nearest Neighbor, KNN)**:<br>
KNN은 전형적인 **게으른 학습기(lazy learner)**이다.<br>
KNN은 훈련 데이터에서 **판별 함수(discriminative function)**를 학습하는 대신 **훈련 데이터셋을 메모리에 저장**하기 때문이다.<br><br>

KNN 알고리즘 3단계 요약
1. 숫자 k와 거리 측정 기준을 선택한다.
2. 분류하려는 샘플에서 k개의 최근접 이웃을 찾는다.
3. 다수결 투표를 통해 클래스 레이블을 할당한다.

이런 **메모리 기반 방식의 분류기**는 수집된 **새로운 훈련 데이터에 즉시 적응**할 수 있는 것이 **장점**이다.<br><br>

단점:<br>
- 새로운 샘플을 분류하는 **계산 복잡도**<br>
- 데이터셋의 **차원(특성)이 적고** 알고리즘이 **효율적인 데이터 구조로 구현되어 있지 않다면** 최악의 경우 훈련 데이터셋의 **샘플 개수에 선형적으로 증가**한다.<br>
- 훈련 단계가 없기 때문에 **훈련 샘플을 버릴 수 없다**.<br>
- 대규모 데이터셋에서 작업하면 **저장 공간에 문제**가 생긴다.<br><br>

#### - **k-최근접 이웃 모델이 학습한 붓꽃 데이터셋의 결정 경계**

```python
# 유클리디안(euclidean) 거리 측정 방식을 사용한 KNN 모델

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, p=2,    # p=2로 지정하면 유클리디안 거리가 되고, p=1로 지정하면 맨해튼 거리가 된다.
                           metric='minkowski')    # minkowski 거리는 유클리디안 거리와 맨해튼(Manhattan) 거리를 일반화한 것
knn.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined,
                      classifier=knn, test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

과대적합과 과소적합 사이에서 올바른 균형을 잡기 위해 **적절한 k를 선택하는 것이 중요**하다.<br>
**실수 값을 가진 특성**에는 보통 **간단한 유클리디안 거리를 사용**한다.<br>
**유클리디안 거리**를 사용하려면 각 특성이 동일하게 취급되도록 **표준화 하는 것이 중요**하다.<br>
