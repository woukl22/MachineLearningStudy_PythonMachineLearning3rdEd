# 3.3.5 사이킷런을 사용하여 로지스틱 회귀 모델 훈련

#### - **사이킷런의 로지스틱 회귀 모델이 만든 결정 경계**

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=100.0, random_state=1)      # 매개변수 'C'에 대해서는 다음 절에서 설명
lr.fit(X_train_std, y_train)

plot_decision_regions(X_combined_std,
                      y_combined,
                      classifier=lr,
                      test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

훈련 샘플이 어떤 클래스에 속할 확률은 predict_proba 메소드를 사용하여 계산한다.

```python
lr.predict_proba(X_test_std[:3, :])
```

첫 번째 행은 첫 번째 붓꽃의 클래스 소속 확률이다.<br>
두 번째 행은 두 번째 붓꽃의 클래스 소속 확률이다.<br>
세 번째 행은 세 번째 붓꽃의 클래스 소속 확률이다.<br><br>

## **샘플 하나의 클래스 레이블을 예측할 때 주의할 점:**<br>
사이킷런은 입력 데이터로 2차원 배열을 기대한다.<br>
따라서 하나의 행을 **2차원 포맷으로 변경**해야 한다.<br>
이 때 사용할 수 있는 **방법** 중 하나는 **넘파이 reshape 메소드**를 사용하여 새로운 차원을 추가하는 것이다.

```python
# 예를 들면 다음과 같다
lr.predict(X_test_std[0, :].reshape(1, -1))
```
