# 3.4 서포트 벡터 머신을 사용한 최대 마진 분류
<br>

#### **서포트 벡터 머신(Support Vector Machine, SVM):**<br>
SVM의 최적화 대상은 **마진을 최대화**하는 것이다.<br>
**마진**은 클래스를 구분하는 초평면(결정 경계)과 이 **초평면에 가장 가까운 훈련 샘플 사이의 거리**로 정의한다.<br>
이런 샘플을 **서포트 벡터(support vector)**라고 한다.

image.png

# 3.4.1 최대 마진

**큰 마진(large margin)**의 결정 경계를 원하는 이유는 **일반화 오차가 낮아지는 경향**이 있기 때문이다.<br>
**작은 마진**의 모델은 **과대적합**하기 쉽다.<br><br>
**마진(margin)**은 **양성 쪽 초평면과 음성 쪽 초평면 사이의 거리**이며, 이것을 최대화하기 위해 SVM을 사용한다.<br>
**SVM의 목적 함수**는 샘플이 정확하게 분류된다는 제약 조건하에서 **2/||w||를 최대화함**으로써 **마진을 최대화**하는 것이다.<br><br>
정리하자면, **모든 음성 클래스 샘플**은 **음성 쪽 초평면 너머**에 있어야 하고, **양성 클래스 샘플**은 **양성 쪽 초평면 너머**에 있어야 한다.<br>
